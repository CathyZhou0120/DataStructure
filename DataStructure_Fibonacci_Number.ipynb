```python
from numpy import empty
import sys
```

```python
def calc_fib(n):
    if n <= 1:
        return(n)
    else:
        arr = empty(n+1,dtype=object)
        arr[0] = 0
        arr[1]=1
        for i in range(2,n+1):
            arr[i]=arr[i-1]+arr[i-2]
        return(arr[n])
        

n = 327305
print(calc_fib(n))
```
